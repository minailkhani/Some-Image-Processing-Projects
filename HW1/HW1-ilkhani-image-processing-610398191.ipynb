#### Computing the difference between values in neighboring pixels:
First printing a part of the image( pixel[0,0] t0 [15,7] )

```python
def print_part_pixel(img, x,y):
    for i in range(x):
        for j in range(y):
            print(img[i,j], end=' ')
        print()
print_part_pixel(img1, 15, 7)
```

now computing **difference** between values in neighboring pixels in **x-direction** and **y-direction** and printing a part of it( pixel[,0] t0 [15,7] )

```python
import numpy as np

diff_x = np.abs(np.diff(img1, axis=1))
diff_y = (np.diff(img1, axis=0))

# print the difference arrays
print("Diff in x-direction:")
print_part_pixel(diff_x, 15, 7)
print("\nDiff in y-direction:")
print_part_pixel(diff_y, 14, 7)
```

### translation
Translation basically means that we are shifting the image by adding/subtracting the X and Y coordinates. WE need this matrix:
$$
\left(\begin{array}{cc} 
1 & 0 & x\\
0 & 1 & y
\end{array}\right)
$$ 

```python
def traslation(img, x,y):
    matrix = np.float32([[1,0,x], 
                         [0,1,y] ])
    return cv2.warpAffine(img2, matrix, (img2.shape[0],img2.shape[1]))


img2Translated = traslation(img2, 50,10)
display_image(img2Translated)
img2Translated = traslation(img2, -10,5)
display_image(img2Translated)
```

#### rotation(90):

```python
height, width = img2.shape[:2]
rotateMatrix = cv2.getRotationMatrix2D(center=(width/2, height/2), angle=90, scale=1)
rotatedImg = cv2.warpAffine(src=img2, M=rotateMatrix, dsize=(width, height))
 
display_image(rotatedImg)
```

```python
img3=cv2.imread(img3Path)
img3 = cv2.cvtColor(img3, cv2.COLOR_BGR2RGB)
display_image(img3)
```

#### Displaying gaussian probability mass function(pmf) and cumulative mass function(CMF) with mu=0 and sigma=1 

```python
from scipy.stats import norm

mu = 0
sigma = 1

x = np.linspace(-5, 5, num=1000)

pdf = norm.pdf(x, mu, sigma)
cdf = norm.cdf(x, mu, sigma)

fig, ax = plt.subplots(1, 2, figsize=(12, 5))
ax[0].plot(x, pdf)
ax[0].set_title('Gaussian Probability Mass Function (PMF)')
ax[1].plot(x, cdf)
ax[1].set_title('Cumulative Mass Function (CMF)')
plt.show()
```

#### Original Histogram:

#### Histogram Equalization:

```python
img = cv2.imread(img3Path)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

equalized = cv2.equalizeHist(gray)

print('Original Image gray')
plt.imshow(gray, cmap='gray')
plt.show()

print('Equalized Image')
plt.imshow(equalized, cmap='gray')
plt.show()

print('Equalized Histogram')
plt.hist(equalized.flatten(), 256, [0, 256])
plt.show()
```

#### Histogram matching:
using the original gray image and equalized image as the reference for matching

```python
matchGray = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8)).apply(gray)

print('matching original image')
plt.imshow(matchGray, cmap='gray')
plt.show()

print('matching original image Histogram')
plt.hist(matchGray.flatten(), 256, [0, 256])
plt.show()


matchEqualized = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8)).apply(equalized)

print('matching eqialized image')
plt.imshow(matchEqualized, cmap='gray')
plt.show()

print('matching eqialized image Histogram')
plt.hist(matchEqualized.flatten(), 256, [0, 256])
plt.show()
```
